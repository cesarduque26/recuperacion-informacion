### Step 3: Implementing Boolean Search
- **Enhance Input Query**: Modify the function to accept complex queries that can include the Boolean operators AND, OR, and NOT.
- **Implement Boolean Logic**:
  - **AND**: The document must contain all the terms. For example, `python AND programming` should return documents containing both "python" and "programming".
  - **OR**: The document can contain any of the terms. For example, `python OR programming` should return documents containing either "python", "programming", or both.
  - **NOT**: The document must not contain the term following NOT. For example, `python NOT snake` should return documents that contain "python" but not "snakg documents.


```python
import pickle
import numpy as np
```

```python
# Cargar el índice invertido desde el archivo
with open('indice_invertido.pkl', 'rb') as archivo:
    indice_invertido_cargado = pickle.load(archivo)
    
# Cargar nombres_archivos desde el archivo
with open('nombres_libros.pkl', 'rb') as archivo:
    nombres_libros = pickle.load(archivo)
```

```python
def buscar_booleano(indice_invertido, consulta):
    consultas = consulta.split()
    resultados = set()
    operador = 'AND' 
    for palabra in consultas:
        if palabra == 'AND':
            operador = 'AND'
        elif palabra == 'OR':
            operador = 'OR'
        elif palabra == 'NOT':
            operador = 'NOT'
        else:
            if operador == 'NOT':
                documentos_con_palabra = indice_invertido.get(palabra, set())
                resultados.difference_update(documentos_con_palabra)
            else:
                documentos_con_palabra = indice_invertido.get(palabra, set())
                if operador == 'AND':
                    resultados = resultados.intersection(documentos_con_palabra)
                elif operador == 'OR':
                    resultados = resultados.union(documentos_con_palabra)
                else:
                    resultados.update(documentos_con_palabra)

    return resultados
```

```python
palabrasUnicas = set()
nombreTextos=[]
contenidoTextos=[]

for texto in textos:
    nombreTexto = re.match(r'^([^:]+):', texto).group(1)
    contenidoTexto = re.sub(r'^([^:]+):', '', texto)
    palabras = re.findall(r'\b(?:[^\W\d_]|(?:[a-zA-Z])+(?:-\w+)*)\b', unidecode(contenidoTexto.lower()))
    nombreTextos.append(nombreTexto)
    contenidoTextos.append(contenidoTexto)
    palabrasUnicas.update(palabras)
palabrasUnicas
```

```python
palabrasUnicas = sorted(list(palabrasUnicas))
```

```python
# Crear archivo CSV
with open('conteo_palabras.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['Texto'] + palabrasUnicas)
    for nombreTexto, contenidoTexto in zip(nombreTextos, contenidoTextos):
        fila = [nombreTexto]
        for palabra in palabrasUnicas:
            palabras = re.findall(r'\b(?:[^\W\d_]|(?:[a-zA-Z])+(?:-\w+)*)\b', unidecode(contenidoTexto.lower()))
            frecuencia = palabras.count(palabra)
            fila.append(frecuencia)
        writer.writerow(fila)

print("Archivo CSV creado exitosamente.")
```

```python
def cargar_csv(nombre_archivo):
    datos = []
    with open(nombre_archivo, newline='') as csvfile:
        reader = csv.reader(csvfile)
        for fila in reader:
            datos.append(fila)
    return datos
```
